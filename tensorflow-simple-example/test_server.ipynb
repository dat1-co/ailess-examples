```python
import requests
import cv2
import base64
import numpy as np
from matplotlib import pyplot as plt
```

```python
r = requests.get('http://localhost:8080/getnum')
resp_json = r.json()
im_enc = resp_json['image']
```

```python
img_bgr = cv2.imdecode(
    np.frombuffer(base64.b64decode(im_enc), dtype=np.uint8), cv2.IMREAD_UNCHANGED
)
img_rgb = cv2.cvtColor(img_bgr,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```
